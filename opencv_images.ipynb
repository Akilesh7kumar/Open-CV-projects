```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(r".\images\input.jpg")
```

```python
#cv2.calcHist(img,channel,mask,histsize,ranges)
#imgs = image source
#channel : index of channel for which we calculate histogram [0]grey [0][1][2]blu,green red
#mask: to mask a image 0 to calculate hist of full image
#histsize:bin count 256
#range : 0 to 256
```

```python
histogram = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
#ravel()flatten image array
plt.hist(img.ravel(),256,[0,256]);
plt.show()
```

```python
#Viewing color componenet
color = ['b','g','r']

for i,col in enumerate(color):
    histogram2 = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.hist(img.ravel(),256,[0,256]);
    plt.plot(histogram2,color =col)
    plt.xlim([0,256])
    plt.show()
    
```

# Drawing shapes and images

```python
image = np.zeros((512,512,3),np.uint8)
image_bw = np.zeros((512,512),np.uint8)
cv2.imshow("black rect color",image)
cv2.imshow("black rect B and W",image_bw)
cv2.waitKey(0)
```

```python
#Drawing a line over a  blacjk square
image = np.zeros((512,512,3),np.uint8)
cv2.line(image,(0,0),(511,511),(255,127,0),5) # x,y,color,pix
cv2.imshow("blue line",image)
cv2.waitKey()
```

```python
#drawing a rectangle\
image = np.zeros((512,512,3),np.uint8)
cv2.rectangle(image,(100,100),(300,250),(127,50,127),5)
cv2.imshow("rectangle",image)
cv2.waitKey()
```

```python
#circles
#cv2.circle(img,center,radius,color,fill)
image = np.zeros((512,512,3),np.uint8)
cv2.circle(image,(350,350),100,(25,64,95),-1)
cv2.imshow("circle",image)
cv2.waitKey()
```

```python
#POLYGON
image = np.zeros((512,512,3),np.uint8)
pts = np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image,[pts],True,(0,0,255),3)
cv2.imshow("circle",image)
cv2.waitKey()
```
